```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The DOS calculated by QUANTUM ESPRESSO is quite rough. We need to smooth it.
# The following link is a nice tutorial that discusses smoothing.
# I used their code to define the fwhm2sigma function.
# https://matthew-brett.github.io/teaching/smoothing_intro.html

def fwhm2sigma(fwhm):
    '''
    Function calculates the standard deviation of a Gaussian curve
    from the full width at half max (fwhm).
        Parameters:
            fwhm (float or int) : full width at half max in meV.
        Returns:
            sigma (float) : standard deviation in meV
    '''
    return fwhm / np.sqrt(8 * np.log(2))

FWHM = 3
```

```python
ph_path = "C:/DFT/Amirhosein/5.1 C2Cn_DOS/"

# conversion from cm^-1 to meV
conv1 = 1.0 / 8.065610
# conversion from cm^-1 to THz
conv2 = 1.0 / 33.356683

e_vals = np.array([]) # phonon energy values
dos_vals = np.array([]) # phonon DOS values

# read DOS files and add values to the arrays
f = open(ph_path+"hBN.dos", "r")
counter = 0
for x in f:
    if counter != 0: # first line in hBN.dos is the table headers which we don't want in our arrays
        e_vals = np.append(e_vals, float(x.split()[0])*conv1) # column #1 is the phonon energy
        dos_vals = np.append(dos_vals, float(x.split()[1]) )  # column #2 is the DOS
    counter += 1
f.close()

# QUANTUM ESPRESSO calculation will likely result in a few negative phonon frequencies.
# There is a lot of information about negative phonon frequencies in the QE forum.
# If the energies are large, then it means that the structure is not stable, but even
# if the structure is stable, calculation errors can still result in negative frequencies.
# These should be small (on the order of meV).
# We correct for these calculation errors by setting the DOS value to zero if the energy <0.
# It doesn't change much, and maybe isn't required, but I did it this way in the results I
# show in the paper. You can comment out the next three lines to see what it looks like without
# this modification.
for i in range(len(e_vals)):
    if e_vals[i] < 0:
        dos_vals[i] = 0
        
# I add a couple zero values to the DOS at high energies. We will apply Gaussian smoothing
# to the DOS, so adding some zeros results in a smooth tail on the right side of the plot
# Again, it doesn't change much, but looks a bit strange without it.
N_append = 30
dos_vals = np.append(dos_vals, np.zeros(N_append))
for i in range(N_append):
    e_vals = np.append(e_vals, e_vals[-1]+(e_vals[2]-e_vals[1]))
    
###################################################################################

# smooth the DOS using Gaussian smoothing.
sigma = fwhm2sigma(FWHM)
smoothed_dos = np.zeros(dos_vals.shape)
for i in range(len(e_vals)):
    kernel = np.exp(-(e_vals - e_vals[i]) ** 2 / (2 * sigma ** 2))
    kernel = kernel / sum(kernel)
    smoothed_dos[i] = sum(dos_vals * kernel)

# plot the DOS
fig = plt.figure(figsize=(13,7))

plt.ylim(0,3.3)
plt.xlim(-2,195)
plt.ylabel('phonon DOS', fontsize=35, labelpad=20)
plt.xlabel('Energy [meV]', fontsize=35, labelpad=20)

plt.fill_between(e_vals, 150*(smoothed_dos/np.sum(smoothed_dos))/(e_vals[2]-e_vals[1]), color='black')
plt.tick_params(labelsize=25)

#plt.savefig('C2Cn_vDOS.png', bbox_inches = 'tight')
plt.show()
```
